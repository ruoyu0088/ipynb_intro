```julia
function uniquelist(list)
    new_list = Set{}()
    for el in list
        push!(new_list, el)
    end
    return collect(new_list)
end

println(uniquelist([1, 2, 1, 3, 2, 3, 4]))
```

```julia
# Function to compute factorial
function factorial(n)
    prod = 1
    for i in 1:n
        prod *= i
    end
    return prod
end

factorial(5)  # returns 120
```

```julia
# Define a Circle type
struct Circle
    radius::Float64
end

# Method to compute area
area(c::Circle) = π * c.radius^2

# Method to compute perimeter
perimeter(c::Circle) = 2 * π * c.radius

# Create an instance
c1 = Circle(5.0)

area(c1), perimeter(c1)
```
