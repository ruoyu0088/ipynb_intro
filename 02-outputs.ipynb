# JupyterLabにおける豊富な出力デモンストレーション

JupyterLabは、単なるPythonコードの実行環境ではなく、さまざまなライブラリの出力を統合的に表示できる強力なインタラクティブ開発環境です。ここでは、JupyterLabで利用できる多様な出力例を紹介し、代表的な可視化ライブラリを通して実際の使い方を解説します。

## Matplotlibによる基本的な2D描画

`matplotlib`は最も基本的で広く使われている可視化ライブラリです。静的なグラフの作成に優れており、JupyterLab上ではノートブックに直接画像として表示されます。

以下の例では、正弦波と余弦波を1つのグラフに描画しています。

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "MS Gothic"
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Matplotlibによる基本的な線グラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

このコードでは、`np.linspace()`で等間隔のx座標を生成し、`plt.plot()`でそれぞれの関数を描画しています。`plt.legend()`を呼び出すことで凡例が自動的に挿入されます。

## Altairによる宣言的データ可視化

`Altair`は「宣言的」可視化ライブラリであり、データと可視化仕様を簡潔に記述できる点が特徴です。JupyterLab上では、インタラクティブな描画をHTMLレンダリングとして表示できます。

```python
import altair as alt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "x": np.linspace(0, 10, 100),
    "y": np.sin(np.linspace(0, 10, 100))
})

chart = alt.Chart(df).mark_line().encode(
    x="x",
    y="y"
).properties(
    title="Altairによるインタラクティブな線グラフ"
).interactive()

chart
```

このコードでは、`mark_line()`で線グラフを指定し、`encode()`でx軸とy軸に対応する列を指定しています。AltairはVega-Liteをバックエンドとして使用しているため、JupyterLab内でズームやホバーが可能です。

## Holoviewsによる動的なデータ探索

`holoviews`は、最小限のコードで豊富なインタラクティブ可視化を実現できるライブラリです。`bokeh`や`matplotlib`など複数のバックエンドを選択できます。

```python
import numpy as np
import holoviews as hv
hv.extension("bokeh", inline=False)

x = np.linspace(0, 10, 200)
y = np.sin(x)
curve = hv.Curve((x, y), "x", "sin(x)").opts(title="Holoviewsによる動的表示", width=500, height=300, show_grid=True)
curve
```

`hv.extension("bokeh")`により、インタラクティブな出力がJupyterLab上に直接レンダリングされます。ホバー時に座標値を確認でき、パンやズームも可能です。

## Plotlyによる高機能インタラクティブプロット

`Plotly`は高度なインタラクティブ機能を備えた可視化ライブラリで、ズーム、回転、ツールチップ表示などを簡単に利用できます。

```python
import plotly.graph_objects as go
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode="lines", name="sin(x)"))
fig.update_layout(
    title="Plotlyによるインタラクティブグラフ",
    xaxis_title="x",
    yaxis_title="y"
)
fig.show()
```

Plotlyの`go.Figure()`オブジェクトは、JupyterLab内で自動的にインタラクティブなHTMLウィジェットとして表示されます。
