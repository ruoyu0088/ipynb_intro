## t-string

```python
name = "World"
times = 100
template = t"Hello {name:}, {times:03d} times"
template
```

```python
from typing import Literal
from string.templatelib import Template, Interpolation

def convert(value: object, conversion: Literal["a", "r", "s"] | None) -> object:
    if conversion == "a":
        return ascii(value)
    elif conversion == "r":
        return repr(value)
    elif conversion == "s":
        return str(value)
    return value

def f(template: Template) -> str:
    parts = []
    for item in template:
        match item:
            case str() as s:
                parts.append(s)
            case Interpolation(value, _, conversion, format_spec):
                value = convert(value, conversion)
                value = format(value, format_spec)
                parts.append(value)
    return "".join(parts)
```

```python
f(template)
```

## interpreters

```python
from concurrent import interpreters

interp = interpreters.create()
interp.exec('''
a = 100
b = 2
c = a + b
alist = [a, b, c]
''')
```

```python
def get_res():
    res = {k:v for k, v in globals().items() if not k.startswith("__")}
    return res
    
interp.call(get_res)
```
