```c++
// Define a simple C++ class
class Circle {
public:
    double radius;

    Circle(double r) : radius(r) {}

    double area() const {
        return 3.141592653589793 * radius * radius;
    }

    double perimeter() const {
        return 2 * 3.141592653589793 * radius;
    }
};
```

```c++
// Create an instance
Circle c1(5.0);

// Access its methods
c1.area()
```

```c++
c1.perimeter()
```
