```rust
struct Point {
    x: f64,
    y: f64,
}

impl Point {
    fn dist(&self) -> f64 {
        (self.x.powi(2) + self.y.powi(2)).sqrt()
    }
}

let p = Point { x: 3.0, y: 4.0 };
p.dist()
```
